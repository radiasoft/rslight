# Visualize SRW datasets

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Final Screen Offset



```python
beam_intensities = np.load('datasets/beam_intensities_watchpoint_position_100runs.npy')
parameters = np.load('datasets/parameters_watchpoint_position_100runs.npy')
print(beam_intensities.shape)
print(parameters.shape)
```

```python
n_plots = 10
idxs = np.random.randint(0, len(beam_intensities), n_plots)
fig, axs = plt.subplots(1, n_plots, figsize=(20,20))

for i, idx in enumerate(idxs):
    axs[i].imshow(beam_intensities[idx].reshape(280, 280))
    axs[i].set_title(np.round(parameters[idx], 5))

plt.show()
```

```python
np.argmin(parameters)
parameters[45]
plt.imshow(beam_intensities[45])
```
